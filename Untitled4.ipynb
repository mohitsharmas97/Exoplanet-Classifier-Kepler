```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Get predictions for each model
y_pred_lr = model.predict(x_test_scaled)
y_pred_rf = model_rf.predict(x_test_scaled)
y_pred_svm = model_svm.predict(x_test_scaled)
y_pred_knn = model_knn.predict(x_test_scaled)

# Create a mapping for the original labels for better visualization
# Assuming the order from LabelEncoder is 0 for CANDIDATE, 1 for CONFIRMED, 2 for FALSE POSITIVE
label_mapping = {0: 'CANDIDATE', 1: 'CONFIRMED', 2: 'FALSE POSITIVE'}
y_test_labels = np.array([label_mapping[i] for i in y_test])
y_pred_lr_labels = np.array([label_mapping[i] for i in y_pred_lr])
y_pred_rf_labels = np.array([label_mapping[i] for i in y_pred_rf])
y_pred_svm_labels = np.array([label_mapping[i] for i in y_pred_svm])
y_pred_knn_labels = np.array([label_mapping[i] for i in y_pred_knn])


# Function to create scatter plot
def plot_actual_vs_predicted(actual_labels, predicted_labels, model_name):
    plt.figure(figsize=(8, 6))
    # Using jitter to better visualize overlapping points
    sns.stripplot(x=actual_labels, y=predicted_labels, jitter=0.2, alpha=0.5)
    plt.title(f'Actual vs. Predicted Disposition ({model_name})')
    plt.xlabel('Actual Disposition')
    plt.ylabel('Predicted Disposition')
    plt.grid(True)
    plt.show()

# Plot for each model
plot_actual_vs_predicted(y_test_labels, y_pred_lr_labels, 'Logistic Regression')
plot_actual_vs_predicted(y_test_labels, y_pred_rf_labels, 'Random Forest')
plot_actual_vs_predicted(y_test_labels, y_pred_svm_labels, 'SVM')
plot_actual_vs_predicted(y_test_labels, y_pred_knn_labels, 'KNN')
```

```python
from sklearn.metrics import confusion_matrix


# Function to plot confusion matrix
def plot_confusion_matrix(actual, predicted, model_name, labels):
    cm = confusion_matrix(actual, predicted)
    plt.figure(figsize=(4,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.title(f'Confusion Matrix ({model_name})')
    plt.xlabel('Predicted Disposition')
    plt.ylabel('Actual Disposition')
    plt.show()

# Plot confusion matrix for each model
plot_confusion_matrix(y_test_labels, y_pred_lr_labels, 'Logistic Regression', label_mapping.values())
plot_confusion_matrix(y_test_labels, y_pred_rf_labels, 'Random Forest', label_mapping.values())
plot_confusion_matrix(y_test_labels, y_pred_svm_labels, 'SVM', label_mapping.values())
plot_confusion_matrix(y_test_labels, y_pred_knn_labels, 'KNN', label_mapping.values())
```
